정답 : 대량의 데이터 처리의 어려움

RDBMS의 대표적인 단점은 아래와 같음
- 성능 : 대용량 데이터 처리, 복잡한 쿼리에 대한 성능이 다른 모델에 비해 낮을 수 있음.
- 확장 어려움: 수직적 확장의 한계 有, 수평적 확장은 복잡하고 비용↑

- 복잡한 조인: 많은 테이블 간의 Join → 성능에 부정적 영향
- 고비용: 서버 하드웨어, 라이센스 비용 상대적으로 ↑

[문제5]

RDBMS의 구성 요소 중 고유한 값을 나타내는 것을 골라주세요

- column
- row
- FK
- PK

정답 : PK

- column(열): 테이블에서 속성(필드)의미. ex) 이름 칼럼에는 중복된 값 존재할수도
- row(행): 테이블에서 1개의 튜플을 나타냄. 전체적으로 고유할 수 있으나 보장하지는 않음
- FK: Foreign Key = 외래 키. 다른 테이블의 기본 키를 참조. 동일한 값이 여러 번 들어올 수 있음.
- PK: Primary Key = 기본 키. 테이블에서 각 행을 유일하게 식별할 수 있으며, 중복 불가 + NULL 불가 조건을 만족해야 함.

[문제10]

데이터 제어 언어(DCL)인 GRANT에 대한 설명으로 옳은 것을 골라주세요

- 사용자에게 특정 작업을 수행할 권한을 부여한다.
- 사용자로부터 특정 작업 수행 권항을 제거한다.
- 다른 사용자에게 동일한 권한을 전달할때 사용한다.
- 부여된 권한 중 일부를 제한할때 사용한다.

정답 :사용자에게 특정 작업을 수행할 권한을 부여한다.


- 사용자로부터 특정 작업 수행 권항을 제거한다 : REVOKE
- 다른 사용자에게 동일한 권한을 전달할때 사용한다 : WITH GRANT OPTION
- 부여된 권한 중 일부를 제한할때 사용한다: 제약조건Constraint와 관련
